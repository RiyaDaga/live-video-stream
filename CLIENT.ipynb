```python
import socket 
import urllib
import pickle
import numpy as np
import cv2

# given url by the mobile app named IP WEBCAM
url = 'http://192.168.29.165:8080/shot.jpg'

# creating the socket
s = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
# Connecting with server
ip= "192.168.29.189"
port=1235
s.connect((ip,port))

while True:
    # recieving the stream video
    x = s.recv(1000000)
    print("Recieved")
    
    imgResp=urllib.request.urlopen(url) # to get the image from the IP camera
    imgNp=np.array(bytearray(imgResp.read()),dtype=np.uint8) # creating array of recevied data
    img=cv2.imdecode(imgNp,-1) # decode the array to usable format
    img = cv2.resize(img,(700,500)) # resizing dispay output window
    ret, buffer = cv2.imencode('.jpg',img) # encoding the data captured by the camera
    bytedata = pickle.dumps(buffer)  # store the buffer data to the bytedata
    # sending the data
    s.send(bytedata)
    try:
        data = pickle.loads(x) # retrieving the recieved video
        data = cv2.imdecode(data,cv2.IMREAD_COLOR) # decoding the video stream in colors
        if data is not None :
            cv2.imshow('caller',data) # displaying the window of videochat
            if cv2.waitKey(10) == 13: # 13 for enter key, exit when enter key is pressed
                break
    except: 
        print("Waiting for the server!") 
     
cv2.destroyAllWindows()
```
